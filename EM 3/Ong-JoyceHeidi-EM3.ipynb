```python
# Clean Month

# dump recweek data
recweek
```

```python
# write code here
abbr_months = [calendar.month_abbr[i] for i in range(1,13)]
def birthmonth_cleaner(birthmonth): 
    if birthmonth.isnumeric(): 
        clean_birthmonth = calendar.month_name[int(birthmonth)]
    elif birthmonth in abbr_months: 
        clean_birthmonth = calendar.month_name[abbr_months.index(birthmonth) + 1]
    else: 
        clean_birthmonth = birthmonth
    return clean_birthmonth

recweek.birthmonth= recweek.birthmonth.map(birthmonth_cleaner)

recweek.head()
```

```python
# Clean Year
# dump recweek data
recweek
```

```python
def year_cleaner(year):
    if len(str(year)) == 2: 
        clean_year = f"19{year}"
    else: 
        clean_year = year 
    return int(clean_year)

recweek.birthyear = recweek.birthyear.map(year_cleaner)
```

```python
# Clean Course
# dump recweek data
recweek
```

```python
conversion = {"BS LM": "BS Legal Management", "Management":"BS Management", "BS CTM": "BS Communications Technology",
            "BS ITE" :"BS IT Entrepreneurship", "Management Engineering" : "BS Management Engineering",
            "CTM": "BS Communications Technology", "BS ME": "BS Management Engineering", "BS MGT": "BS Management", 
            "Information Technology Entrepreneurship": "BS IT Entrepreneurship", "LM": "BS Legal Management", 
            "ITE": "BS IT Entrepreneurship", "MGT" : "BS Management", "ME" : "BS Management Engineering", 
            "Communications Technology": "BS Communications Technology", "Legal Management": "BS Legal Management"}

recweek.course = recweek.course.replace(conversion)

    
```

```python
# mobile number cleaning

# dump recweek data
recweek
```

```python
# write code here
recweek.phonenumber = recweek.phonenumber.str.replace(r"^(\+63 )*9","09")
```

```python
# add genders
# hint: load "studentprofiles.csv" as new dataframe and add to our dataframe using Pandas merge(...) function
# dump recweek data
#final_recweek
```

```python
# write code here
student_profile = pd.read_csv("studentprofiles.csv")
final_recweek = recweek.merge(student_profile)[["firstname", "lastname", "gender", "birthmonth", "birthday", "birthyear", "course","yearlevel", "phonenumber"]]
```

```python
# write code here
df2 = final_recweek
def carrier_identifier(phonenumber): 
    if phonenumber[:4] == "0918": 
        carrier = "Smart"
    else: 
        carrier = "Globe"
    return carrier 

df2["carrier"] = df2.phonenumber.map(carrier_identifier)

df2.groupby(["carrier","gender"]).count().rename(columns= {"firstname":"count"})["count"].reset_index()
```
